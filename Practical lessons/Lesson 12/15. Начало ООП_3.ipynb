# Задача 16

**Написать собственный класс ```Scan``` (```Скан```).**

**В качестве необходимых атрибутов класс ```Scan``` должен имплементировать:**

* Уникальное имя скана ```name```;
* Атрибут ```_points``` содержащий коллекцию объектов ```ScanPoint```;
* Атрибут ```_borders``` содержащий в себе данные о границах скана ($x_{min}, x_{max}, y_{min}, y_{max}, \dots$);

**Переопределите стандартный метод приведения объекта к строке ```__str__(self)``` и ```__repr__(self)``` (методы должны вернуть объект строки ```str```, описывающий объект в котором он вызван).**

**Переопределите стандартный метод ```__len__(self)``` который определяет логику работы стандартной функции ```len()```. Метод должен возвращать количество точек, содержащихся в скане.**

**Переопределите стандартный метод ```__iter(self)___``` который определяет логику работы стандартной функции ```iter()```. Метод должен возвращать объект итератора для последовательного перебора коллекции ```_points```.**

**Напишите метод ```scan_parser(self, scan_parser_obj, file_path)```, где:**
* ```scan_parser_obj``` - объект определяющий логику загрузки точек из файла в скан (например ```ScanTXTParser``` - объект который будет загружать точки из текстового файла);
* ```file_path``` - путь до файла с загружаемыми данными.
  

**Напишите метод ```scan_plotter(self, scan_plotter_obj, *args, **kwargs)```, где:**
* ```scan_plotter_obj``` - объект определяющий логику визуализации скана (например ```ScanPlotterMPL``` - объект который будет отрисовыать скан используя пакет `matplotlib`);
* ```*args```, ```**kwargs``` - дополнительные аргументы, необходимые для работы объекта ```scan_plotter_obj```.
  
**Напишите метод ```scan_filter(self, scan_filter_obj, *args, replace=True, **kwargs)```, где:**
* ```scan_filter_obj``` - объект определяющий логику фильтрации или разрежения исходного скана (например ```ScanDecimator``` - объект который будет разрежать скан, исходя из заданного процента точек, который необходимо оставить в скане или ```ScanPolygonFilter``` - который оставит в скане только те точки, которые находятся внутри заданной полилинии и тп);
* ```replace``` - определяет, заменяются ли данные в исходном объекте (если ```True```) или (если ```False```) метод возвращает новый объект класса ```Scan``` (имя скана ```name``` можно сгенерировать через агрегацию имени исходного скана и суффикс, определяемый способом фильтрации (например ```scan_1``` в котором через ```ScanDecimator``` осталось $10\%$ точек можно назвать ```scan_1_decim_10``` и тп) саму логику возвращения нового скана можно реализовать через изменение копии исходного скана (через метод ```copy()``` или ```deepcopy()```;
* ```*args```, ```**kwargs``` - дополнительные аргументы, необходимые для работы объекта ```scan_filter_obj```.

> Постарайтесь придумать и реализовать максимальное количество способов фильтрации сканов для разных условий. (Например оставляющие только точки в определенном радиусе от заданной точки или оставляющие только точки определенного цвета (не конкретного значения оттенка, а именно цвета), и тп)
